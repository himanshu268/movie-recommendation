```python
import pandas as pd
import ast  
import numpy as np
```

```python
data1=pd.read_csv('tmdb_5000_credits.csv')
data2=pd.read_csv('tmdb_5000_movies.csv')
```

```python
data1.head(1)
# data1.shape
```

```python
data2.head(1)
```

```python
movies=data1.merge(data2,on='title')
```

```python
movies.info()
```

```python
movies=movies[['movie_id','title','cast','crew','genres','overview','keywords']]
```

```python
movies=movies.dropna()
```

```python
movies.isnull().sum()
```

```python
movies.title
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
        
```

```python
movies['cast']=movies['cast'].apply(convert)
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
def convert1(obj):
    L=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
            L.append(i['name'])
            count+=1
        else:
            break
    return L
        
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
def direc(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew']=movies['crew'].apply(direc)
```

```python
movies.head(1)
```

```python
def coll(l):
    l1=[]
    for i in l:
        l1.append(i.replace(" ",""))
    return l1
```

```python
movies['cast']=movies['cast'].apply(coll)
movies['crew']=movies['crew'].apply(coll)
movies['genres']=movies['genres'].apply(coll)
movies['keywords']=movies['keywords'].apply(coll)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['tag']=movies['cast']+movies['crew']+movies['genres']+movies['keywords']+movies['overview']
```

```python
movies['tag']=movies['tag'].apply(lambda x:" ".join(x))
```

```python
movies.head(1)
```

```python
movies = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
#used for finding the most frequent items
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
cv.get_feature_names()
```

```python
vector=cv.fit_transform(movies['tag']).toarray()
```

```python
#used for find the similarity between the movies on the basis of vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = movies[movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1]) #used for finding similarity 
    for i in distances[1:6]:
        print(movies.iloc[i[0]].title)
```

```python
recommend('Newlyweds')
```
